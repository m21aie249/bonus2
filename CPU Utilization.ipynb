```python
%matplotlib notebook
frame_len=200
y=[]
fig=plt.figure(figsize=(8,3))
def animate(i):
    y.append(cpu_percent())
    if(len(y)<=frame_len):
        plt.cla()
        plt.plot(y,'r',label="Real time CPU usage")
    plt.tight_layout()
    
ani=FuncAnimation(plt.gcf(),animate,interval=1000)

    
```
